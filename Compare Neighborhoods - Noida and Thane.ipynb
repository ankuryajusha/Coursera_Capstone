### Introduction

#### We are now comparing the neighborhoods in Noida and Thane.
#### To do this we will be use the foresquare data generated during exploring both cities.
#### Data will be clustered using k-means clustering and we will show the top 10 venue categories for the neighborhoods
#### for each cluster. 

```python
# Import required libraries

import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```python
# Get data generated peviously during exploring cities

df_thane = pd.read_csv("thane_res_sectors_foresquare_data_2000m.txt")
df_thane.drop(columns='Unnamed: 0', axis=1, inplace=True)
df_noida = pd.read_csv("noida_res_sectors_foresquare_data_2000m.txt")
df_noida.drop(columns='Unnamed: 0', axis=1, inplace=True)
```

```python
df_thane
```

```python
df_noida
```

```python
# Adding string 'Noida' to neighborhood for city Noida

df_noida['Neighborhood'] = df_noida['Neighborhood'] + ', Noida'
df_noida
```

```python
# Merging the data for Noida and Thane

df = df_noida.append(df_thane, ignore_index=True)
df
```

```python
# Count the number of venue categories for each neighborhood 

df.groupby('Neighborhood').count()
```

```python
# Count the unique venue categories

len(df['Venue Category'].unique())
```

```python
# one hot encoding
df_onehot = pd.get_dummies(df['Venue Category'])

# add neighborhood column to the dataframe
df_onehot['Neighborhood'] = df['Neighborhood']

# move neighborhood column to the first column
first_column = df_onehot.pop('Neighborhood')
df_onehot.insert(0, 'Neighborhood', first_column)

df_onehot.head()
```

```python
df_grouped = df_onehot.groupby('Neighborhood').mean().reset_index()
df_grouped
```

```python
k = 10

df_grouped_clustering = df_grouped.drop('Neighborhood', axis=1)

# run k-means clustering
kmeans = KMeans(n_clusters=k, random_state=0)
kmeans.fit(df_grouped_clustering)
```

```python
np.unique(kmeans.labels_, return_counts=True)
```

```python
# Function to get top venues

def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = df_grouped['Neighborhood']

for ind in np.arange(df_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(df_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

```python
neighborhoods_venues_sorted.shape
```

```python
# Add cluster labels to dataframe

neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
neighborhoods_venues_sorted.head()
```

```python
# Get the data generated by geolocator for Noida and Thane

df_neighborhood_noida = pd.read_csv("noida_res_sectors_geolocator_data.txt")
df_neighborhood_noida.drop(columns='Unnamed: 0', axis=1, inplace=True)
df_neighborhood_noida.rename(columns={'Sector':'Neighborhood'}, inplace=True)
df_neighborhood_noida['Neighborhood'] = df_neighborhood_noida['Neighborhood'] + ', Noida'

df_neighborhood_thane = pd.read_csv("thane_res_sectors_geolocator_data.txt")
df_neighborhood_thane.drop(columns='Unnamed: 0', axis=1, inplace=True)
df_neighborhood_thane.rename(columns={'Neighbourhood':'Neighborhood'}, inplace=True)
```

```python
# Merge the geolocator data

df_neighborhood_merged = df_neighborhood_noida.append(df_neighborhood_thane, ignore_index=True)
```

```python
# Join the geolocator data with data generated after clustering

df_merged = df_neighborhood_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')
df_merged
```

```python
# Function to generate unrelated clusters

def uniqueNeighborhoods(df):
    
    df_final = df
    df_final = df_final.iloc[0:0]

    for cluster in np.sort(df['Cluster Labels'].unique()):

        df1 = df[df['Cluster Labels'] == int(cluster)]

        city_list = ['Noida', 'Thane', 'Navi Mumbai']
        x = 0
        for value in city_list:
            if len(df1[df1['Neighborhood'].str.contains(value)]) > 0:
                x=x+1

        if x>1:
            pass
        else:
            df_final = df_final.append(df1)

    return(df_final)
```

```python
unrelated_neighborhoods = uniqueNeighborhoods(df_merged)
unrelated_neighborhoods
```
